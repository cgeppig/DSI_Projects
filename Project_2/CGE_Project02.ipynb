##### Using Pandas' built in `melt` function, pivot the weekly ranking data to be long rather than wide.
```python
billboardMelt = billboardMelt.dropna()
billboardMelt
```

```python
number1 = billboardMelt[billboardMelt['Rank'] == 1]
number1.head()
```

```python
number1pivot = pd.pivot_table(number1, index=['artist.inverted', 'track'],
                             values=['Rank'],
                             aggfunc=(len))
```

```python
billboardMeltPivot = pd.pivot_table(billboardMelt, index=['artist.inverted', 'track', 'time', 'genre', 'week1'], 
               values=['Rank'], aggfunc=(len, np.mean, min))

billboardMeltPivot = billboardMeltPivot.reset_index()

billboardMeltPivot.head()
```
